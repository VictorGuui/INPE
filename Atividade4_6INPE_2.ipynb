```python
estados = gpd.read_file("/content/drive/MyDrive/BR_UF_2022.shx")
```

```python
dsetAnomalia = xr.open_dataset("/content/drive/MyDrive/Climatologia_AmericaDoSul_CHIRPS.nc")
```

```python
lat = dsetAnomalia.latitude.values
lon = dsetAnomalia.longitude.values
var = dsetAnomalia['precip']
```

```python
dsetAnomalia
```

```python
precipitacao = dsetAnomalia['precip']
```

```python
precipitacao = precipitacao.sel(time=slice('1991-01-01', '2020-12-01'))
```

```python
precipitacao['time.season']
```

```python
media_climatologica = precipitacao.groupby('time.month').mean(dim='time')
```

```python
media_climatologica
```

```python
ano = 1998
meses = range(1, 13)

anomalias_sazonais = []

for mes in meses:
    anomalia_sazonal = var.sel(time=f'{ano}-{mes:02d}') - media_climatologica.sel(month=mes)
    anomalias_sazonais.append(anomalia_sazonal)
```

```python
# Configuração da projeção do mapa e sua extensão geográfica
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent((-90.0, -30.0, -60.0, 15.0))
ax.coastlines(resolution='110m', color='black')
ax.add_feature(cfeature.BORDERS, linewidth=0.5, edgecolor='black')

# Configuração das grades do mapa
gl = ax.gridlines(draw_labels=True)
gl.xlines = False
gl.ylines = False
gl.right_labels = False
gl.top_labels = False

# Define os níveis de contorno
levels = np.linspace(-120.0, 110.0, 11)

# Inicializa um gráfico de contorno preenchido com os dados de anomalia
cnplot = ax.contourf(anomalias_sazonais.longitude, anomalias_sazonais.latitude, anomalias_sazonais, cmap='BrBG', levels=levels, extend='both')

# Adiciona uma barra de cores (colorbar) na parte inferior do gráfico
cbar = plt.colorbar(cnplot, orientation='horizontal', pad=0.07, shrink=0.6)
cbar.set_label('Precipitação de Anomalia (mm/day) \n Janeiro 2022')
estados.plot(ax=ax, color='none', edgecolor='black')

# Salva a figura em um arquivo chamado 'anomalias_sazonais.png' com uma resolução de 300 dpi
plt.savefig('anomalias_sazonais.png', dpi=300)
```

```python
# Assuming "anomalias_sazonais" is your array
for idx, data_array in enumerate(anomalias_sazonais):
    # Extracting data and coordinates
    precipitation_data = data_array.values.squeeze()
    latitude = data_array.latitude.values
    longitude = data_array.longitude.values
    time = data_array.time.values  # Remove [0] to get the array of time values

    ax.add_feature(cfeature.BORDERS, linewidth=0.5, edgecolor='black')

    # Creating a meshgrid for plotting
    lon, lat = np.meshgrid(longitude, latitude)
    levels = np.linspace(-120.0, 110.0, 11)

    # Creating a plot
    plt.figure(figsize=(10, 6))
    plt.contourf(lon, lat, precipitation_data, cmap='BrBG', levels=levels, extend='both')
    plt.colorbar(label='Precipitation (mm)')
    plt.title(f'Anomalias de precipitação  - {pd.to_datetime(time[0]).strftime("%Y-%m")}')  # Use time[0] for the title
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    estados.plot(ax=ax, color='none', edgecolor='black')

    plt.show()
```

```python
# Assuming "anomalias_sazonais" is your array
num_items = len(anomalias_sazonais)
num_rows = 4
num_cols = 3

fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 12), subplot_kw={'projection': ccrs.PlateCarree()})

for idx, data_array in enumerate(anomalias_sazonais):
    # Extracting data and coordinates
    precipitation_data = data_array.values.squeeze()
    latitude = data_array.latitude.values
    longitude = data_array.longitude.values
    time = data_array.time.values[0]

    # Creating a meshgrid for plotting
    lon, lat = np.meshgrid(longitude, latitude)

    # Creating a subplot
    row_idx = idx // num_cols
    col_idx = idx % num_cols
    ax = axs[row_idx, col_idx]

    levels = np.linspace(-120.0, 110.0, 11)

    # Adding contour around South America
    ax.add_feature(cfeature.BORDERS, linewidth=0.5, edgecolor='black')

    # Plotting on the current subplot
    im = ax.contourf(lon, lat, precipitation_data, cmap='BrBG',levels=levels, extend='both')
    ax.set_title(f'Anomalias de precipitação - {pd.to_datetime(time).strftime("%Y-%m")}')
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    estados.plot(ax=ax, color='none', edgecolor='black')


# Adjust layout for better spacing
plt.tight_layout()

# Add a colorbar using the last plot
cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])
fig.colorbar(im, cax=cbar_ax, label='Precipitation (mm)')

plt.show()
```
